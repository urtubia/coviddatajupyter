```python
import pandas as pd


df = pd.read_csv("https://covidtracking.com/api/v1/states/daily.csv")

df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')

# df.head()

# df.describe()


#df['state']

df[df['state'] == 'NV']
nv_df = df[df['state'] == 'NV']
nv_df
```

```python
import matplotlib.pyplot as plt

nv_positives = nv_df[['date','positiveIncrease', 'totalTestResultsIncrease']]

nv_positives.plot(x='date', y=['positiveIncrease','totalTestResultsIncrease'])
```

```python
ax = nv_positives.plot(x='date',
                      y=['positiveIncrease','totalTestResultsIncrease'],
                      title='Daily Cases and Daily Tests',
                      subplots=True,
                      figsize=(12,8))
```
